```python
# import des libraries pour le prophet
import warnings
import pandas as pd

import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
 
import seaborn as sns
sns.set()
sns.set(rc={'figure.figsize':(14.7,10.27)})
 
from plotly import tools
import plotly.graph_objs as go
import gc
 
from datetime import datetime
 
# prophet model 
from fbprophet import Prophet
# prophet preformance
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
from prophet.plot import plot_cross_validation_metric
from IPython.display import set_matplotlib_formats
```

```python
# On crÃ©e des listes vides avec les features qui seront ulisÃ© pour prÃ©dire si nous devons utiliser l'algorithme 
my_nbr_of_crypto_tweets_list = []
my_nbr_of_nasdaq_tweets_list = []
my_Volatility_Gold_Close_list = []
my_Volatility_Gold_Vol_list = []
my_Volatility_Oil_Close_list = []
my_Volatility_Oil_Vol_list = []
my_Volatility_Gas_Close_list = []
my_Range_Gold_Close_list = []
my_Range_Gold_Vol_list = []
my_Range_Gas_Close_list = []
my_Range_Oil_Vol_list = []
my_Mean_PMI_list = []
my_Mean_FED_list = []
my_Mean_Unemploy_list = []
my_Volatility_VIX_Close_list = []
my_Range_VIX_Close_list = []
my_mean_Volatility_Close_list = []
my_mean_Volatility_Vol_list = []
my_mean_Range_Close_list = []
my_mean_Range_Vol_list = []
my_start_date_List = []
```

```python
# On dÃ©fini les fcontion utilisÃ© pour faire nos features (same as feature creator)
def volatility(data):
    # On crÃ©e une fonction qui retourne le volatiltÃ© de l'action dans une window dÃ©finie plus bas
    return np.log(data/data.shift()).std()
def max_range(data):
    # On crÃ©e une fonction qui retourne l'amplitude max de la valeure
    return np.max(data)-np.min(data)
```

```python
# On dÃ©fini le debut de la periode car prophet prends ~1mois pour predire 
start_date = datetime(2022, 1, 31)
# On crÃ©e un DF vide qui viendra Ãªtre update avec les 7 valeurs et on viendra faire nos calculs 
df_features_crypto_forecasted = pd.DataFrame()
for i in range(0, df_features_crypto[start_date:].shape[0]//7):
    print(i, '/', df_features_crypto[start_date:].shape[0]//7)
    # On set notre window (7) comme lors de l'entrainement
    start_date_i = start_date + timedelta(days = i*7)
    df_forecast_full = pd.DataFrame([])
    for e in df_features_crypto.columns : 
            df_model= df_features_crypto[df_features_crypto.index[0]+ timedelta(days = i*7): start_date_i].reset_index().rename(columns={'Date':'ds', e:'y'})[['ds', 'y']]
            # CrÃ©ation du modÃ¨le prophet
            prophet_model = Prophet()
            prophet_model.fit(df_model)
            # PrÃ©vision sur window jours
            future = prophet_model.make_future_dataframe(periods=3, freq='D')
            # Ajout d'un titre
            forecast = prophet_model.predict(df=future)
            forecast = forecast.rename(columns = {'yhat': e, 'ds' : 'Date'})
            forecast = forecast[['Date', e]].tail(3)
            #On rÃ©cupÃ¨re les 3 derniÃ¨res valeurs du prophet
            forecast = forecast.set_index('Date')
            #display(forecast)
            #fig1 = prophet_model.plot(forecast)
            df_forecast_full[e] = forecast
    df_features_crypto[df_features_crypto.index[0]+ timedelta(days = i*2): start_date+timedelta(days = i*7)]
    df_features_crypto_forecasted = pd.concat([df_features_crypto[start_date+timedelta(days = i*7)-timedelta(days = 3): start_date+timedelta(days = i*7)], df_forecast_full], axis =0)[-7:]
    print(df_features_crypto_forecasted)
    list_Vol_Volatility = []
    list_Clo_Volatility = []
    list_Vol_Range = []
    list_Clo_Range = []
    # Si on se trouve sur une col utilisÃ© dans une feature features on calcule cette derniÃ¨re
    for j in df_features_crypto_forecasted.columns:
        if j == 'nbr_of_cryptos_tweets':
            my_nbr_of_crypto_tweets = int(sum(df_features_crypto_forecasted[f'{j}']))
        elif j =='nbr_of_nasdaq_tweets':
            my_nbr_of_nasdaq_tweets = int(sum(df_features_crypto_forecasted[f'{j}']))
        elif j == 'Gold_Close':
            my_Volatility_Gold_Close = volatility(df_features_crypto_forecasted[f'{j}'])
            my_Range_Gold_Close = max_range(df_features_crypto_forecasted[f'{j}'])
        elif j == 'Gold_Volume':
            my_Volatility_Gold_Vol = volatility(df_features_crypto_forecasted[f'{j}'])
            my_Range_Gold_Vol = max_range(df_features_crypto_forecasted[f'{j}'])
        elif j == 'Oil_Close':
            my_Volatility_Oil_Close = volatility(df_features_crypto_forecasted[f'{j}'])
        elif j == 'Oil_Volume':
            my_Volatility_Oil_Vol = volatility(df_features_crypto_forecasted[f'{j}'])
            my_Range_Oil_Vol = max_range(df_features_crypto_forecasted[f'{j}'])
        elif j == 'Gas_Close':
            my_Volatility_Gas_Close = volatility(df_features_crypto_forecasted[f'{j}'])
            my_Range_Gas_Close = max_range(df_features_crypto_forecasted[f'{j}'])
        elif j =='PMI':
            my_Mean_PMI = np.mean(df_features_crypto_forecasted[f'{j}'])
        elif j =='Unemploy':
            my_Mean_Unemploy = np.mean(df_features_crypto_forecasted[f'{j}'])
        elif j =='FED':
            my_Mean_FED = np.mean(df_features_crypto_forecasted[f'{j}'])
        elif 'Vol' in j:
            list_Vol_Volatility.append(df_features_crypto_forecasted[f'{j}'])
            list_Vol_Range.append(df_features_crypto_forecasted[f'{j}'])
        elif 'Close' in j:
            list_Clo_Volatility.append(df_features_crypto_forecasted[f'{j}'])
            list_Clo_Range.append(df_features_crypto_forecasted[f'{j}'])
        else: 
            my_Volatility_VIX_Close = volatility(df_features_crypto_forecasted[f'{j}'])
            my_Range_VIX_Close = max_range(df_features_crypto_forecasted[f'{j}'])
        
    list_vol_Volatility_i = []
    list_clo_Volatility_i = []
    list_vol_Range_i = []
    list_clo_Range_i = []
    for c in list_Vol_Volatility:
        list_vol_Volatility_i.append(volatility(c))
    for d in list_Clo_Volatility:
        list_clo_Volatility_i.append(volatility(d))
    for e in list_Vol_Range:
        list_vol_Range_i.append(max_range(e))
    for f in list_Clo_Range:
        list_clo_Range_i.append(max_range(f))
    # On fait la moyennes des volatilitÃ©s des volumes et des closes de chaques actions
    my_mean_Volatility_Close = np.mean(list_clo_Volatility_i)
    my_mean_Volatility_Vol = np.mean(list_vol_Volatility_i)
    my_mean_Range_Close = np.mean(list_clo_Range_i)
    my_mean_Range_Vol = np.mean(list_vol_Range_i)
    # On ajoute les valeurs calculÃ©s dans les listes
    my_nbr_of_crypto_tweets_list.append(my_nbr_of_crypto_tweets)
    my_nbr_of_nasdaq_tweets_list.append(my_nbr_of_nasdaq_tweets)
    my_Volatility_Gold_Close_list.append(my_Volatility_Gold_Close)
    my_Volatility_Gold_Vol_list.append(my_Volatility_Gold_Vol)
    my_Volatility_Oil_Close_list.append(my_Volatility_Oil_Close)
    my_Volatility_Oil_Vol_list.append(my_Volatility_Oil_Vol)
    my_Volatility_Gas_Close_list.append(my_Volatility_Gas_Close)
    my_Range_Gold_Close_list.append(my_Range_Gold_Close)
    my_Range_Gold_Vol_list.append(my_Range_Gold_Vol)
    my_Range_Oil_Vol_list.append(my_Range_Oil_Vol)
    my_Range_Gas_Close_list.append(my_Range_Gas_Close)
    my_Mean_PMI_list.append(my_Mean_PMI)
    my_Mean_FED_list.append(my_Mean_FED)
    my_Mean_Unemploy_list.append(my_Mean_Unemploy)
    my_Volatility_VIX_Close_list.append(my_Volatility_VIX_Close)
    my_Range_VIX_Close_list.append(my_Range_VIX_Close)
    my_mean_Volatility_Close_list.append(my_mean_Volatility_Close)
    my_mean_Volatility_Vol_list.append(my_mean_Volatility_Vol)
    my_mean_Range_Close_list.append(my_mean_Range_Close)
    my_mean_Range_Vol_list.append(my_mean_Range_Vol)
    my_start_date_List.append(start_date_i)
    
```

```python
# On crÃ©e le df des features utilisÃ© avec en plus la start date 
df_features_to_ML_crypto = pd.DataFrame({
    "close_volatility" : my_mean_Volatility_Close_list,'close_range': my_mean_Range_Close_list, "vol_volatility" : my_mean_Volatility_Vol_list,'vol_range': my_mean_Range_Vol_list, "nbr_of_cryptos_tweets" : my_nbr_of_crypto_tweets_list,
    'nbr_of_nasdaq_tweets' : my_nbr_of_nasdaq_tweets_list, "Volatility_Gold_Close" : my_Volatility_Gold_Close_list, "Volatility_Gold_Vol" : my_Volatility_Gold_Vol_list, "Volatility_Oil_Close": my_Volatility_Oil_Close_list,
    "Volatility_Oil_Vol" : my_Volatility_Oil_Vol_list, "Volatility_Gas_Close"  : my_Volatility_Gas_Close_list, "Range_Gold_Close" : my_Range_Gold_Close_list, 
    "Range_Gold_Vol" : my_Range_Gold_Vol_list, "Range_Oil_Vol":my_Range_Oil_Vol_list,"Range_Gas_Close" : my_Range_Gas_Close_list,'Mean_FED': my_Mean_FED_list, 'Mean_PMI': my_Mean_PMI_list, "Mean_Unemploy" : my_Mean_Unemploy_list,
    "Volatility_VIX": my_Volatility_VIX_Close_list, 'Range_VIX': my_Range_VIX_Close_list, "start_date": my_start_date_List
})
df_features_to_ML_crypto
```
