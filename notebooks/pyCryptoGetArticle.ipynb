```python
from newspaper import Article
from googlesearch import search
import pandas as pd
```

```python
from googlesearch import search
search("Google", start=0, stop=10)
```

```python
class getArticles:
    
    def __init__(self, words_search = 'market crypto news' , after= '2023-01-01' , before='2022-01-01', start=0, stop=20):
         self._words_search = words_search
         self._after = after
         self._before = before
         self._start = start
         self._stop = stop
         self._list_url_article = []
         self._articles = [] 
        
    def get_links(self):
        self._list_url_article= []
        query = "market crypto news after:" + self._after + " before:" + self._before
        for url in search(query, lang='en', start=self._start, stop=self._stop):
            print(url)
            self._list_url_article.append(url)
        
    def get_articles(self):
        for url in self._list_url_article:
         try:
          article = Article(url)
          article.download()
          article.parse()

          title = article.title
          link = article.url
          authors = article.authors
          date = article.publish_date
          text = article.text
          print('**********************************')
          print(f'Title: {title}')
          print(f'Link: {link}')
          print(f'Author: {authors}')
          print(f'Publish Date: {date}')
          print(f'Text: {text}')
          print('**********************************')
          self._articles.append([date, title, authors, text])
         except :
                continue   

        df = pd.DataFrame(self._articles, columns=['Date', 'Title', 'Authors', 'Text'])
        return df
        
```

```python
#Test value default
test_getArticles = getArticles()
test_getArticles.get_links()
df = test_getArticles.get_articles()
display(df.head(5))
df.to_csv('../data/articles.csv', sep='\t', encoding='utf-8')
```
