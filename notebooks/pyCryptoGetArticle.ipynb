```python
from newspaper import Article
from googlesearch import search
import pandas as pd
import datetime
```

```python
class getArticles:
    
    def __init__(self, words_search = 'market news' , after= '2018-01-01' , before='2018-12-31', start=0, stop=20):
         self._words_search = words_search
         self._after = after
         self._before = before
         self._start = start
         self._stop = stop
         self._list_url_article = []
         self._articles = [] 
        
    def get_links(self):
        self._list_url_article= []
        query = "market crypto news after:" + self._after + " before:" + self._before
        for url in search(query, lang='en', start=self._start, stop=self._stop):
            print(url)
            self._list_url_article.append(url)
        
    def get_articles(self):
        for url in self._list_url_article:
         try:
          article = Article(url)
          article.download()
          article.parse()

          title = article.title
          link = article.url
          authors = article.authors
          date = article.publish_date
          text = article.text
          print('**********************************')
          print(f'Title: {title}')
          print(f'Link: {link}')
          print(f'Author: {authors}')
          print(f'Publish Date: {date}')
          print(f'Text: {text}')
          print('**********************************')
          self._articles.append([date, title, authors, text])
         except :
                continue   

        df = pd.DataFrame(self._articles, columns=['Date', 'Title', 'Authors', 'Text'])
        return df
        
```

```python
#Test value default
start_date = "2022-02-24"
end_date = "2022-04-22"
test_getArticles = getArticles(after = start_date , before = end_date)
test_getArticles.get_links()
df = test_getArticles.get_articles()
display(df.head(5))
```

```python
df_clean = df.dropna(axis = 0, how ='any')
display(df_clean)
display(df_clean.isnull())
```
