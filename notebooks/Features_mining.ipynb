```python
def Fill_nas(serie, end):
    for i in df.index.year.unique():
        for j in df.index.month.unique():
            #print(df['PMI'][f'{j}-{i}'].unique())
            for k in serie[f'{j}-{i}'].unique():
                if math.isnan(k) == False:
                    serie[f'{j}-{i}'] = k
            if f'{j}-{i}' == end:
                break
Fill_nas(df['PMI'], '8-2022')
Fill_nas(df['FED'], '7-2022')
Fill_nas(df['Michi'], '6-2022')
Fill_nas(df['Unemploy'], '8-2022')
```
